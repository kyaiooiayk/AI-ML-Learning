# Series decomposition

<div class="alert alert-block alert-info">
<font color=black><br>

- Time series decomposition involves thinking of a series as a combination of level, trend, seasonality, and noise components. 
- This means the time series can be split and the sole trend highlighted.

<br></font>
</div>

# Import modules

```python
from pandas import read_csv
from random import randrange
from matplotlib import pyplot, rcParams
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib as mpl

rcParams['figure.figsize'] = 15, 5
rcParams['font.size'] = 20
```

# Create the dataset

```python
series = [i+randrange(10) for i in range(1,100)]
len(series)
```

```python
plt.plot(range(len(series)), series, lw = 3)
pyplot.show()
```

# Additive Decomposition

<div class="alert alert-block alert-info">
<font color=black><br>

- Running the example creates the series, performs the decomposition, and plots the 4 resulting series. 
- We can see that the entire series was taken as the trend component and that there was no seasonality.
- This is a good example where the naive, or classical, decomposition was -->>NOT<<-- able to separate the noise that we added from the linear trend. 
- The naive decomposition method is a simple one, and there are more advanced decompositions available, like Seasonal and Trend decomposition using Loess or STL decomposition.
- Caution and healthy skepticism is needed when using automated decomposition methods. 

<br></font>
</div>

```python
result = seasonal_decompose(series, model = 'additive', period = 1) 
rcParams['font.size'] = 15
rcParams['figure.figsize'] = 10, 10
result.plot()
pyplot.show()
```

# Multiplicative Decomposition

<div class="alert alert-block alert-info">
<font color=black><br>

- Exponential changes can be made linear by data transforms. 
- In this case, a quadratic trend can be made linear by taking the square root. 
- An exponential growth in seasonality may be made linear by taking the natural logarithm.

<br></font>
</div>

```python
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', period = 1) 
rcParams['font.size'] = 15
rcParams['figure.figsize'] = 10, 10
result.plot()
pyplot.show()
```

# Airline Passengers Dataset

<div class="alert alert-block alert-info">
<font color=black><br>

- Running the example plots the observed, trend, seasonal, and residual time series. 
- We can see that the trend and seasonality information extracted from the series does seem reasonable. 
- The residuals are also interesting, showing periods of high variability in the early and later years of the series.

<br></font>
</div>

```python
series = read_csv('../DATASETS/airline-passengers.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
# We assume a multiplicative model!
result = seasonal_decompose(series, model = 'multiplicative') 
rcParams['font.size'] = 15
rcParams['figure.figsize'] = 10, 10
result.plot()
pyplot.show()
```
